```python
import numpy as np
import pandas as pd
import pickle
import cv2
from collections import defaultdict
```

```python
with open('statesToData.pickle', 'rb') as handle:
    statesToData = pickle.load(handle)
```

```python
statesToData["CA"].CENSUS_ID
```

```python
standard_file_name_list = ['road_0.tif', 'road_1.tif', 'building_0.tif', 'building_1.tif', 'USGSImageryWMS.jpg']
num_channels = len(standard_file_name_list)
```

```python

folder_new_id_list = set()
for key in statesToData:
    for census_id in statesToData[key].CENSUS_ID:
        folder_new_id_list.add(census_id)
print(len(folder_new_id_list))
```

```python

folder_old_id_list = []
for id_ in folder_new_id_list:
    if id_[0] == '0':
        folder_old_id_list.append(id_[1:])
    else:
        folder_old_id_list.append(id_) # match up to the folder names...
```

```python
file_size = (200, 200) # Use a smaller size of images...Otherwise the laptop cannot work at a certain point.
image_array = np.zeros((len(folder_old_id_list), num_channels+2, file_size[0], file_size[1])) # default: black images
```

```python
census_id_to_index = {}
empty_images_folder_names = []
```

```python
#Xiao hu needs to collect more blackwhite images 'road_0.tif', 'road_1.tif', 'building_0.tif', 'building_1.tif'
import os
empty = 0
colored = 0
not_empty = 0
for folder_name_idx in np.arange(len(folder_old_id_list)):
    if folder_name_idx%100==0:
        print(folder_name_idx)
    for file_name_idx in np.arange(len(standard_file_name_list)):
        full_file_name = os.path.join("/home/jtl/Dropbox (MIT)/project_demand_unstructured_data/project_1_nhts_auto_cnn/data/img/", folder_old_id_list[folder_name_idx], standard_file_name_list[file_name_idx])
        img = cv2.imread(full_file_name, cv2.IMREAD_UNCHANGED)
        if img is None:
            #print("empty")
            
            #print(standard_file_name_list[file_name_idx])
            empty += 1
            empty_images_folder_names.append(full_file_name)
            pass # default is black images...so when images do not exist. Keep the zeros
        elif (standard_file_name_list[file_name_idx] == "USGSImageryWMS.jpg"):
            #print("colored")
            colored += 1
            image_array[folder_name_idx, file_name_idx, :, :] = cv2.resize(img[:,:,0], file_size).astype(np.uint8)
            
            image_array[folder_name_idx, file_name_idx+1, :, :] = cv2.resize(img[:,:,0], file_size).astype(np.uint8)
            
            image_array[folder_name_idx, file_name_idx+2, :, :] = cv2.resize(img[:,:,0], file_size).astype(np.uint8)
            
            census_id_to_index[str(folder_old_id_list[folder_name_idx])] = folder_name_idx
        else:
            #print("not empty")
            not_empty += 1
            image_array[folder_name_idx, file_name_idx, :, :] = cv2.resize(img, file_size).astype(np.uint8)
            
            census_id_to_index[str(folder_old_id_list[folder_name_idx])] = folder_name_idx
            
print("empty: ",empty)
print("colored: ",colored)
print("not empty: ", not_empty)
```
