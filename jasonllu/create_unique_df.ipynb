```python
import pickle
import pandas as pd
import numpy as np
```

```python
with open('../statesToData.pickle', 'rb') as handle:
    statesToData = pickle.load(handle)
```

```python
all_data_list = []
```

```python
for i,state in enumerate(statesToData):
    df = statesToData[state]
    all_data_list.append(df)
```

```python
all_data = pd.concat(all_data_list, axis = 0)
```

```python
all_data.shape
```

```python
all_data.head()
```

```python
data_unique_census_id = all_data.drop_duplicates(subset = "CENSUS_ID", inplace = False)
```

```python
print(all_data.shape)
print(data_unique_census_id.shape)
```

```python
data_unique_census_id.head()
```

```python
data_unique_census_id = data_unique_census_id[data_unique_census_id.columns.tolist()[::-1]]
```

```python
data_unique_census_id.head(10)
```

```python
print(data_unique_census_id.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_unique, test_unique = train_test_split(data_unique_census_id, train_size = 5/6, test_size = 1/6, shuffle = True)
```
