```python
stateToNumSamples = {key:int(round((len(statesToData[key])/sum([len(statesToData[key]) for key in statesToData])),3)*100000) for key in statesToData}
```

```python
stateToNumSamples
```

```python
sum(stateToNumSamples.values())
```

```python
stratified_data = [statesToData[key].sample(stateToNumSamples[key]+stateToNumSamples[key]//10,replace=False) for key in statesToData]
```

```python
stratified_data[0][:680].shape
```

```python
train_list = []
test_list = []
```

```python
for i,key in enumerate(statesToData):
    df = stratified_data[i]
    print(df.shape)
    train_num = stateToNumSamples[key]
    test_num = stateToNumSamples[key]//10
    train_list.append(df[:train_num])
    test_list.append(df[train_num:])
    print(df[:train_num].shape)
    print(df[train_num:].shape)
```

```python
print(len(train_list))
print(len(test_list))
```

```python
stratified_train_df = pd.concat(train_list,axis=0)
stratified_test_df = pd.concat(test_list,axis=0)
```

```python
print(stratified_train_df.shape)
print(stratified_test_df.shape)
```

```python
stratified_train_df.to_csv("stratified_train100k.csv")
```

```python
stratified_test_df.to_csv("stratified_test100k.csv")
```
